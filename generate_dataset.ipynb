```python
import json
import random

# 🔥 Conversation templates
user_starts = [
    "Hey", "Hi", "Hello", "Yo", "What's up?",
    "Are you free?", "What are you doing?", "How are you?"
]

assistant_replies = [
    "Hey!", "Hi there!", "Hello!", "Not much", "Just chilling",
    "Doing good", "What's up?", "All good here"
]

followups = [
    "I'm bored", "Let's do something", "Wanna go out?",
    "I feel tired", "I have a lot of work",
    "Thinking to watch a movie", "Any plans?",
    "I'm hungry", "Feeling stressed"
]

# 🔥 Reply styles
reply_styles = [
    ["Sounds good!", "I'm in!", "Let's do it!"],
    ["What time?", "Where should we go?", "Any ideas?"],
    ["Same here", "Yeah it's boring", "We should do something"],
    ["Take some rest", "You should relax", "Sounds like a long day"],
    ["Maybe watch something", "Let's go out", "Want to chill?"],
    ["Order something", "Let's grab food", "Cook something?"],
    ["Don't worry", "You'll be fine", "It happens"]
]

# 🔥 Generate one sample
def generate_sample():
    u1 = random.choice(user_starts)
    a1 = random.choice(assistant_replies)
    u2 = random.choice(followups)

    conversation = f"User: {u1}\nAssistant: {a1}\nUser: {u2}"

    replies = random.choice(reply_styles)

    output = "\n".join(replies)

    return {
        "instruction": "Generate 3 smart reply suggestions.",
        "input": conversation,
        "output": output
    }

# 🔥 Generate dataset
data = [generate_sample() for _ in range(1000)]

# 🔥 Save to JSONL
with open("dataset_1000.jsonl", "w", encoding="utf-8") as f:
    for item in data:
        f.write(json.dumps(item) + "\n")

print("✅ dataset_1000.jsonl created with 1000 samples")
```

```python
import json

fixed_data = []

with open("dataset_1000.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        item = json.loads(line)

        replies = item["output"].split("\n")

        # ensure 3 replies
        while len(replies) < 3:
            replies.append("...")

        item["output"] = f"""Reply 1: {replies[0]}
Reply 2: {replies[1]}
Reply 3: {replies[2]}"""

        fixed_data.append(item)

# save fixed dataset
with open("dataset_fixed.jsonl", "w", encoding="utf-8") as f:
    for item in fixed_data:
        f.write(json.dumps(item) + "\n")

print("✅ Fixed dataset saved as dataset_fixed.jsonl")
```
